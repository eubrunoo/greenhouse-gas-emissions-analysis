```python
each_gas_emission = years_emissions.groupby('Gás')[['Emissão']].sum(numeric_only = True).sort_values('Emissão', ascending= False)
```

```python
each_gas_emission.plot(kind= 'barh', figsize= (10, 6))
```

```python
each_gas_emission.iloc[0:9]
```

```python
transformation = ((each_gas_emission.iloc[0:9].sum(numeric_only = True))/(each_gas_emission.sum(numeric_only = True)).iloc[0] * 100)
```

```python
print(f'The emission of CO2 matches with {float(transformation.iloc[0]):.2f}% of the total gas emission in Brazil from 1970 to 2021')
```
